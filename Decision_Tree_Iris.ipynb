```python
from sklearn import tree
```

```python
X = [[0, 0], [1, 1]]
```

```python
Y = [0, 1]
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(X, Y)
```

```python
clf.predict([[2., 2.]])
```

```python
clf.predict_proba([[2., 2.]])
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier() #objek clf dari class tree
clf = clf.fit(X, y) #pelatihan
```
