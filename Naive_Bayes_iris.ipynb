```python
from sklearn.datasets import load_iris  #import dataset
from sklearn.model_selection import train_test_split  #import split
from sklearn.naive_bayes import GaussianNB #nb model gaussianNB
X, y = load_iris(return_X_y=True)  #
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)  #split test 0,3 atau 30% dan train 0,7 atau 70%
gnb = GaussianNB()  #bikin objek namanya gnb isinya algoritna gaussiannb
y_pred = gnb.fit(X_train, y_train).predict(X_test)  #pelatihan gnb.fit # uji .predict
print("Number of mislabeled points out of a total %d points : %d"  #print menampilkan
      % (X_test.shape[0] , #shape untuk menghitung dimensi garis
      (y_test != y_pred).sum()))
```
